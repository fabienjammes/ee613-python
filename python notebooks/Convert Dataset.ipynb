```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
```

#### Convert from .mat to .npy 

```python
from os import listdir
```

```python
ROOT = 'data/2Dletters/'
ROOT_NEW = 'python_data/2Dletters/'
filenames = listdir(ROOT)
```

```python
for filename in filenames:
    raw_data = loadmat(ROOT + filename)['demos'][0]
    data = []
    for data_i in raw_data:
        data.append(data_i[0][0][0])
    data = np.array(data)
    f = open(ROOT_NEW + filename[:-4] + '.npy','wb')
    np.save(f,data)
    f.close()
```

### Try to visualize one data 

```python
filenames = listdir(ROOT_NEW)
data = np.load(ROOT_NEW + filenames[3])
```

```python
for data_i in data:
    plt.plot(data_i[0,:], data_i[1,:])
```
