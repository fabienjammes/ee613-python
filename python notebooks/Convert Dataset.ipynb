```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

#### Convert from .mat to .npy 

```python
from os import listdir
```

```python
ROOT = 'data/2Dletters/'
ROOT_NEW = 'python_data/2Dletters/'
filenames = listdir(ROOT)
```

```python
for filename in filenames:
    raw_data = loadmat(ROOT + filename)['demos'][0]
    data = dict()
    data['pos'] = []
    data['vel'] = []
    data['acc'] = []
    for data_i in raw_data:
        data['pos'].append(data_i[0][0][0])
        data['vel'].append(data_i[0][0][1])
        data['acc'].append(data_i[0][0][2])

    data = np.array(data)
    f = open(ROOT_NEW + filename[:-4] + '.pkl','wb')
    pickle.dump(data,f)
    f.close()
```
